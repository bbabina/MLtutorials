```python
import pandas as pd
df = pd.read_csv("salaries.csv")
df.head()
```

```python
inputs = df.drop("salary_more_then_100k", axis = 'columns')
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['column_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs.head()
```

```python
input_n = inputs.drop(['company', 'job', 'degree'], axis = 'columns')
input_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(input_n,target)
```

```python
model.score(input_n,target)
```

```python
model.predict([[2,2,1]]) # google sales executive
```

```python
model.predict([[2,0,1]]) # google business manager
```
