```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.01
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum (x*(y-y_predicted))
        bd = -(2/n)*sum (y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = m_curr - learning_rate * bd
        ## print("m {}, b {}, cost {} iteration {}".format(m_curr, b_curr, cost, i))
  

```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```

```python
import pandas as pd
df = pd.read_csv("homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
```

```python
merged = pd.concat([df, dummies], axis = 'columns')
merged
```

```python
# always drop one of the dummy variable, this may create dummy variable trap
#when one variable can be derived from other variable this is called multicolinear and this may create  dummy variable trap 
```

```python
final = merged.drop(['town', 'west windsor'], axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('price', axis = 'columns')
X
```

```python
y = final.price
y
#dependent variable-Y 
```

```python
model.fit(X, y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```
