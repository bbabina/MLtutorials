```python
from sklearn import svm, datasets
iris = datasets.load_iris()
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df[47:52]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size= 0.3)
```

```python
model = svm.SVC(kernel='rbf', C=30, gamma='auto')
model.fit(X_train, y_train)
model.score(X_test, y_test)
# A model selction but as you run score changes 
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(svm.SVC(kernel='linear', C=10, gamma='auto'), iris.data, iris.target)  
# can be many values
```

```python
cross_val_score(svm.SVC(kernel='rbf', C=10, gamma='auto'), iris.data, iris.target)  #same as the for loop does
```

```python
cross_val_score(svm.SVC(kernel='rbf', C=20, gamma='auto'), iris.data, iris.target)  #same as the for loop does
```

```python
import numpy as np
kernels=['rbf', 'linear']
C = [1,10,20]
avg_scores = {}
for kval in kernels:
    for cval in C:
        cv_scores = cross_val_score(svm.SVC(kernel=kval, C=cval, gamma='auto'),iris.data, iris.target, cv=5)
        avg_scores[kval + '_' + str(cval)] = np.average(cv_scores)
avg_scores                
```
